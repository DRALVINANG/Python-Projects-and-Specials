#Loading the Data

```python
from sklearn.datasets import load_iris

iris = load_iris()
```



---



#Method 1: Using Pandas + Colab's Dataframe

```python
import pandas as pd

df = pd.DataFrame(iris.items(), columns=["Key", "Value"])

display(df)
```



---



#Method 2: Using Self Created Dictionary

```python
def print_neat_dictionary(dictionary):
  for key, value in dictionary.items():
    print("#-----------------------------------------------------------------------------#")
    print(f'\n{key.upper()}:\n\n{value}\n')
```

```python
print_neat_dictionary(iris)
```
