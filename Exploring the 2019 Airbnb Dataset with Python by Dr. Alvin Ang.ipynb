##What is the distribution of listing prices?

```python
sns.distplot(listings_df['price'])
listings_df.price.mean()
```



---



##How many listings are more expensive than 500$/night?
Could we remove those outliers from the visualization? What technique can we use?


```python
affordable_df = listings_df[listings_df['price'] <= 500]
sns.distplot(affordable_df['price'])
affordable_df.price.mean()
```



---



##What is the distribution of listing prices based on the Neighbourhood Group?

We can use a Violinplot:

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(15, 8))

sns.violinplot(data=affordable_df, x="neighbourhood_group", y="price")
```

##Can we plot the listings on a map?

Back to matplotlib, we can use a scatterplot:

```python
affordable_df.plot(
    kind='scatter',
    x='longitude',
    y='latitude',
    c='price',
    cmap='inferno',
    colorbar=True,
    alpha=0.8,
    figsize=(12,8))
```

##Adding NYC map underneath!

```python
import urllib.request
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import io

background_image_url = 'https://raw.githubusercontent.com/lewagon/data-images/master/workshops/Neighbourhoods_New_York_City_Map.png'

# Open the URL and read the image data.
with urllib.request.urlopen(background_image_url) as url_obj:
    image_data = url_obj.read()

# Wrap the image data in a io.BytesIO object.
image_file = io.BytesIO(image_data)

# Convert the image data to a NumPy array.
background_image = np.array(Image.open(image_file))

# Plot the background image.
plt.imshow(background_image, zorder=0, extent=[-74.258, -73.7, 40.49, 40.92])

# Plot the affordable housing data.
ax = plt.gca()
affordable_df.plot(
    ax=ax,
    zorder=1,
    kind='scatter',
    x='longitude',
    y='latitude',
    c='price',
    cmap='inferno',
    colorbar=True,
    alpha=0.8,
    figsize=(12,8)
)

plt.show()
```
