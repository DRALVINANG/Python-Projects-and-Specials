#2) np.random.randn

Generates random samples from a normal distribution with a mean of 0 and a standard deviation of 1

```python
arr = np.random.randn(7,4)
#7 rows 4 columns of random data (mean 0, SD 1)

arr
```

```python
sd = np.std(arr)
sd

#to find out the SD of the arr
```

#5) Creating 2 Clusters of Random Numbers

```python
import pandas as pd
import numpy as np
```

```python
a1 = list(-np.random.random(50))
b1 = list(-np.random.random(50))

df1 = {
    'a' : a1,
    'b' : b1
}

df1 = pd.DataFrame(df1)
df1.head()
```

```python
a2 = list(np.random.random(50))
b2 = list(np.random.random(50))

df2 = {
    'a' : a2,
    'b' : b2
}

df2 = pd.DataFrame(df2)
df2.head()
```

```python
df3 = pd.concat([df1, df2])
df3
```

```python
import matplotlib.pyplot as plt
plt.scatter(df3.a, df3.b)
```
