#Create DataFrame

```python
import pandas as pd

# Create a sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Age': [25, 30, 35, 40, 45],
        'City': ['New York', 'London', 'Paris', 'Tokyo', 'Sydney']}
df = pd.DataFrame(data)
df
```



---



#METHOD 1: LOC and ILOC

##LOC

```python
df.set_index('Name', inplace=True)
df
```

```python
df.loc['Alice']
```

```python
df.iloc[0]
```



---



```python
df.reset_index(inplace=True)
df
```

```python
df.loc[
    df['City'] == 'Paris']
```

```python
df.loc[
    df['Age'] > 30,
      ['Name', 'City']]
```



---



##ILOC

```python
df.iloc[1:4,
        0:2]
```

```python
df.iloc[-1,
        0:2]
```



---



#METHOD 2: QUERY

```python
# Filter rows where Age is greater than 30
df.query('Age > 30')
```



---



#METHOD 3: Logical Indexing

```python
# Filter rows where Age is greater than 30 using logical indexing

df[df['Age'] > 30]
```



---



#METHOD 4: Lambda and APPLY

```python
# Filter rows where the length of the Name is greater than 4 using lambda

df[df['Name'].apply(lambda x: len(x) > 4)]
```



---



#METHOD 5: Set and Intersection then ISIN()

```python
# Convert the 'Name' column to a set
names_set = set(df['Name'])

# Define a set of names you want to filter
filter_names = {'Alice', 'Charlie', 'Eve'}

# Use set intersection to find the common elements
filtered_names = names_set.intersection(filter_names)

# Filter the DataFrame based on the filtered names
filtered_df = df[df['Name'].isin(filtered_names)]

print(filtered_df)
```



---



#METHOD 6: STR.startswith()

```python
# Filter rows where Name starts with 'A'

df[df['Name'].str.startswith('A')]
```



---



#METHOD 7: FILTER

```python
df.filter(items=['Age', 'City'])
```
